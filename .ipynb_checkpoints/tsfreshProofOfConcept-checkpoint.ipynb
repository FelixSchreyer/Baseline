```python
import datetime

# Cleans Dataframe by removing the seperate columns for hour, minute, 
# second and milisecond and replacing them by a singe datetime column
def create_time_column(df_time):
    datetime.time()
    custom_header_names = ['hour', 'minute', 'second', 'milisecond', 'x', 'y']
    
    df_time.columns.name = 'column_names'
    df_time.columns = custom_header_names
    default_year = 2023
    default_month = 1
    default_day = 1

    df_time.reset_index(drop=True)

    # Create a new datetime column
    df_time['time_column'] = pd.to_datetime({
        'year': default_year,
        'month': default_month,
        'day': default_day,
        'hour': df_time['hour'],
        'minute': df_time['minute'],
        'second': df_time['second'],
        'millisecond': df_time['milisecond']
    })
    columns_to_drop = ['hour', 'minute', 'second', 'milisecond']
    df_time = df_time.drop(columns=columns_to_drop)
    
    new_column_order = ['time_column', 'x', 'y']
    df_time = df_time[new_column_order]
    
    return df_time

# Iterates through dataframe to add a RUL for every row
def create_rul_columns(df_rul):
    df_rul['RUL'] = df_rul.apply(remaining_useful_lifetime, axis=1)
    df_rul['integer_column'] = df_rul['RUL'].apply(map_to_integer)
    return df_rul

# Calculates RUL by computing time difference between last sensor values and input entry
def remaining_useful_lifetime(df_rul):
    current = df_rul['time_column']
    beginning = combined_df['time_column'].min()
    failure = combined_df['time_column'].max()
    dateTimeDifference = failure - current
    dateTimeDifferenceInMinutes = dateTimeDifference.total_seconds() / 60
    result = ((failure-current)/(failure-beginning))
    return result

def map_to_integer(value):
    return int(value * 5)

def add_id_column(df, value):
    df['id'] = value
    return df
```

```python
import os

# Directory containing CSV files
csv_directory = './Data/'

# List to store DataFrames from each CSV
dataframes = []

# Loop through each file in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_directory, filename)
        df = pd.read_csv(file_path, header=None, skiprows=lambda x: x % 75)
        dataframes.append(df)

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df = create_time_column(combined_df)
combined_df = create_rul_columns(combined_df)
combined_df = add_id_column(combined_df, "1")
```

## Extract Features
Based on the Dataset we first set the extraction settings to subsequently extract the features and filter them regarding their impact. Extraction and Filtering can be done in the same step. 

```python
# We are very explicit here and specify the `default_fc_parameters`. If you remove this argument,
# the ComprehensiveFCParameters (= all feature calculators) will also be used as default.
# Have a look into the documentation (https://tsfresh.readthedocs.io/en/latest/text/feature_extraction_settings.html)
# or one of the other notebooks to learn more about this.
extraction_settings = ComprehensiveFCParameters()

X = extract_features(df, column_id='id', column_sort='time', 
                     default_fc_parameters=extraction_settings,
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
```

```python
X.head()
```

```python
X_filtered = select_features(X, y)
```

## Train Model
After obtaining the extracted and filtered features, they can now be used to train the machine learning model. THis is where adapt comes into play for transfer learning. 

```python
from adapt.utils import make_classification_da
from adapt.feature_based import FA
```

```python
X_full_train, X_full_test, y_train, y_test = train_test_split(X, y, test_size=.4)
X_filtered_train, X_filtered_test = X_full_train[X_filtered.columns], X_full_test[X_filtered.columns]
```

```python
classifier_full = DecisionTreeClassifier()
classifier_full.fit(X_full_train, y_train)
print(classification_report(y_test, classifier_full.predict(X_full_test)))
```

```python
Xs, Xt, ys, yt = train_test_split(X, y, test_size=.4)
TL_model = FA(classifier_full, Xt=Xt, yt=yt, random_state=0)
TL_model.fit(Xs, ys)
print(classification_report(y_test, TL_model.predict(X_full_test)))
```

```python
classifier_filtered = DecisionTreeClassifier()
classifier_filtered.fit(X_filtered_train, y_train)
print(classification_report(y_test, classifier_filtered.predict(X_filtered_test)))
```
