```python
# Directory containing CSV files
csv_directory = '../FEMTO/Bearing1_1/'

# List to store DataFrames from each CSV
dataframes = []

# Loop through each file in the directory
for filename in os.listdir(csv_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(csv_directory, filename)
        df = pd.read_csv(file_path, header=None, skiprows=lambda x: x % 375)
        dataframes.append(df)

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
# Add Time Column, RUL columns and ID Column
combined_df = create_time_column(combined_df)
#combined_df = create_rul_columns(combined_df)
#combined_df = create_id_column(combined_df, "1")
#combined_df = combined_df.sort_values(by='time_column')
#combined_df = combined_df.reset_index(drop=True)
combined_df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(combined_df['time_column'], combined_df['x'], label='X')
plt.plot(combined_df['time_column'], combined_df['y'], label='Y')
plt.xlabel('Time')
plt.ylabel('X/Y')
plt.title('Time Series Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Create rolled dataframe with at least 1000 observations
df_rolled = roll_time_series(combined_df, column_id='id', column_sort="time_column", 
                             min_timeshift=1000, rolling_direction=200)
df_rolled
```

```python
df_rolled = create_rul_columns(df_rolled)
df_rolled
```

```python
df_rolled['y'] = df_rolled['y'].fillna(method='ffill')
```

```python


# Create features based upon rolled dataframe
X = extract_features(df_rolled.drop(columns=["RUL", "integer_column"]), column_id='id', column_sort='time_column', 
                     column_kind=None, column_value=None,
                     default_fc_parameters= tsfresh.feature_extraction.MinimalFCParameters(),
                     # we impute = remove all NaN features automatically
                     impute_function=impute)
X.head()
```

```python

combined_y=df_rolled['integer_column']
combined_y.index = df_rolled['id'].tolist()

distinct_entries_series = combined_y.groupby(level=0).first()
distinct_entries_series
```

```python
X_train_selected = select_features(X, distinct_entries_series)
X_train_selected
```

## Begin of FA and Model Training

```python
Xs = X_train_selected
Xt = X_train_selected
ys = distinct_entries_series
yt = distinct_entries_series
```

```python
model = FA(RandomForestClassifier(), Xt=Xt[:10], yt=yt[:10], random_state=0)
```

```python
model.fit(Xs, ys)


model.score(Xt, yt)
print(classification_report(yt, model.predict(Xt)))
```
